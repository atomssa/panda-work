```python
from math import *
from matplotlib import *
import numpy as np
import matplotlib.pyplot as plt
import scipy.constants as const
from scipy.integrate import quad

%pylab inline
```

```python
def fun(x, A):
    return 1 + A*cos(x)*cos(x)
def fun0(x):
    return (1 + cos(x)*cos(x))/1.5
def fun1(x):
    return (1 + 0.4*cos(x)*cos(x))/1.2
```

```python
int1 = quad(fun, 0, pi, args=(1.0))[0]/pi;
int2 = quad(fun, 0, pi, args=(0.4))[0]/pi;
print "int1 = ",int1
print "int2 = ",int2

int1 = quad(fun0, 0, pi)[0];
int2 = quad(fun1, 0, pi)[0];

print "int1 = ",int1
print "int2 = ",int2
```

```python
x = np.linspace(0.0,pi,100)
y0 = map(fun0, x)
y1 = map(fun1, x)
fig, ax = plt.subplots(figsize=(8,8))
ax.plot(x, y0, 'b')
ax.plot(x, y1, 'k')
plt.axis([0,pi,0,1.5])
print plt.axis()
#plt.yticks(np.arange(0, 5, 10.0))
plt.show()
```
